# Tumor Mutational Burden and Survival in Colorectal Cancer
### A Survival Analysis using MSK-CHORD Dataset

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter
import numpy as np
from lifelines import CoxPHFitter
import matplotlib.ticker as mticker
from lifelines.statistics import logrank_test
import warnings
warnings.filterwarnings("ignore")
# load data
df = pd.read_csv('/Users/haiyuanyu/Desktop/629project/final/msk_chord_2024_clinical_data.tsv', sep="\t")

# view 
print(df.shape)
print(df.columns.tolist())
df.head()
```

```python
df_crc = df[df["Cancer Type"] == "Colorectal Cancer"].copy()
print("CRC shape:", df_crc.shape)
print(df_crc["Cancer Type"].value_counts(dropna=False))
```

```python
df_crc["OS_months"] = pd.to_numeric(df_crc["Overall Survival (Months)"], errors="coerce")

df_crc["event"] = df_crc["Overall Survival Status"].astype(str).apply(
    lambda x: 1 if "DECEASED" in x else 0
)

df_crc["TMB"] = pd.to_numeric(df_crc["TMB (nonsynonymous)"], errors="coerce")
df_crc["mutation_count"] = pd.to_numeric(df_crc["Mutation Count"], errors="coerce")
df_crc["age"] = pd.to_numeric(df_crc["Current Age"], errors="coerce")

df_crc["sex"] = df_crc["Sex"].astype("category")
df_crc["race"] = df_crc["Race"].astype("category")
df_crc["stage"] = df_crc["Stage (Highest Recorded)"].astype("category")
```

```python
print(df_crc[["OS_months", "event", "TMB", "mutation_count", "age"]].isna().sum())
print(df_crc["TMB"].describe())
print(df_crc["Stage (Highest Recorded)"].value_counts(dropna=False))
```

```python
median_tmb = df_crc["TMB"].median()
df_crc["TMB_group"] = df_crc["TMB"].apply(
    lambda x: "High" if pd.notna(x) and x >= median_tmb else
              ("Low" if pd.notna(x) else pd.NA)
)

print(df_crc["TMB_group"].value_counts(dropna=False))
```

```python


# cleaning race 
df_crc["Race_clean"] = df_crc["Race"].replace({
    "No Value Entered": pd.NA,
    "Unknown": pd.NA,
    "Pt Refused To Answer": pd.NA
})

# log transformation
df_crc["log_TMB"] = np.log1p(df_crc["TMB"])

cox_df2 = df_crc[[
    "OS_months", "event", "log_TMB",
    "Current Age", "Sex", "Race_clean", "Stage (Highest Recorded)"
]].dropna().copy()

cox_df2 = pd.get_dummies(
    cox_df2,
    columns=["Sex", "Race_clean", "Stage (Highest Recorded)"],
    drop_first=True
)

cph2 = CoxPHFitter()
cph2.fit(cox_df2, duration_col="OS_months", event_col="event")

res2 = cph2.summary[[
    "exp(coef)", "exp(coef) lower 95%", "exp(coef) upper 95%", "p"
]]
print(res2)
```

Higher TMB is associated with lower hazard of death (HR < 1), after adjusting for age, sex, race, and stage.

```python
cph2.check_assumptions(cox_df2, p_value_threshold=0.05, show_plots=False)
```

```python


plot_df = df_crc.dropna(subset=["OS_months", "event", "TMB_group"]).copy()

# log-rank
low = plot_df[plot_df["TMB_group"] == "Low"]
high = plot_df[plot_df["TMB_group"] == "High"]
res = logrank_test(low["OS_months"], high["OS_months"],
                   event_observed_A=low["event"],
                   event_observed_B=high["event"])

plt.figure(figsize=(8,6))
kmf = KaplanMeierFitter()

for group, color in zip(["Low","High"], ["#1f77b4","#d62728"]):
    m = plot_df["TMB_group"] == group
    kmf.fit(plot_df.loc[m,"OS_months"], plot_df.loc[m,"event"], label=f"{group} TMB")
    sf = kmf.survival_function_
    plt.step(sf.index, sf.iloc[:,0], where="post", color=color, label=f"{group} TMB", linewidth=2)

plt.title("Overall Survival by TMB Group (CRC)")
plt.xlabel("Months")
plt.ylabel("Survival Probability")
plt.text(0.78, 0.2, f"log-rank p = {res.p_value:.3f}", transform=plt.gca().transAxes)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

KM curves show no significant difference in survival between TMB groups (log-rank p = 0.23).

```python
cox_df3 = cox_df2.copy()
# log_TMB × Stage_4 interaction
for col in cox_df3.columns:
    if "Stage (Highest Recorded)_Stage 4" in col:
        cox_df3["logTMB_x_Stage4"] = cox_df3["log_TMB"] * cox_df3[col]

cph3 = CoxPHFitter()
cph3.fit(cox_df3, duration_col="OS_months", event_col="event")
print(cph3.summary.loc[["log_TMB","logTMB_x_Stage4"]])
```

```python
cox_df_strata = df_crc[[
    "OS_months", "event", "log_TMB",
    "Current Age", "Sex", "Race_clean", "Stage (Highest Recorded)"
]].dropna().copy()

cox_df_strata = pd.get_dummies(
    cox_df_strata,
    columns=["Sex", "Race_clean"],
    drop_first=True
)

cph_strata = CoxPHFitter()
cph_strata.fit(
    cox_df_strata,
    duration_col="OS_months",
    event_col="event",
    strata=["Stage (Highest Recorded)"]
)

print(cph_strata.summary[[
    "exp(coef)", "exp(coef) lower 95%", "exp(coef) upper 95%", "p"
]])
```

```python
res_plot = res2.copy().reset_index()
res_plot = res_plot.rename(columns={"covariate": "variable"})

# only keep key variables
keep_vars = [
    "log_TMB",
    "Current Age",
    "Sex_Male",
    "Race_clean_Black Or African American",
    "Race_clean_White",
    "Stage (Highest Recorded)_Stage 4"
]

res_plot = res_plot[res_plot["variable"].isin(keep_vars)].copy()

# rename labels
label_map = {
    "log_TMB": "log(TMB)",
    "Current Age": "Age",
    "Sex_Male": "Male vs Female",
    "Race_clean_Black Or African American": "Black (vs Others)",
    "Race_clean_White": "White (vs Others)",
    "Stage (Highest Recorded)_Stage 4": "Stage IV vs I–III"
}

# create label column first
res_plot["label"] = res_plot["variable"].map(label_map)

# set custom order
order = [
    "log(TMB)",
    "Stage IV vs I–III",
    "Age",
    "Male vs Female",
    "Black (vs Others)",
    "White (vs Others)"
]

res_plot["label"] = pd.Categorical(
    res_plot["label"],
    categories=order,
    ordered=True
)

# sort by custom order
res_plot = res_plot.sort_values("label", ascending=False)

y_pos = np.arange(len(res_plot))

plt.figure(figsize=(7, 5))

# points and CI
plt.errorbar(
    res_plot["exp(coef)"],
    y_pos,
    xerr=[
        res_plot["exp(coef)"] - res_plot["exp(coef) lower 95%"],
        res_plot["exp(coef) upper 95%"] - res_plot["exp(coef)"]
    ],
    fmt='o', markersize=6,
    color='black',
    ecolor='black',
    capsize=3
)

# middle line HR=1
plt.axvline(1, color='gray', linestyle='--')

# Y axis
plt.yticks(y_pos, res_plot["label"])

# log scale for x label
plt.xscale("log")

plt.xlabel("Hazard Ratio (log scale)")
plt.title("Adjusted Hazard Ratios (Cox Model)")

for spine in ["top", "right"]:
    plt.gca().spines[spine].set_visible(False)

plt.tight_layout()
plt.savefig("forest_main.png", dpi=300)
plt.show()
```

```python
print(df_crc["Stage (Highest Recorded)"].value_counts(dropna=False))
```

```python
df_crc["stage4_bin"] = df_crc["Stage (Highest Recorded)"].apply(
    lambda x: "Stage IV" if str(x) == "Stage 4" else "Non-Stage IV"
)
print(df_crc["stage4_bin"].value_counts(dropna=False))
```

```python


# 2. Run subgroup-specific Cox models
subgroup_results = []

for subgroup in ["Non-Stage IV", "Stage IV"]:
    sub = df_crc[df_crc["stage4_bin"] == subgroup].copy()

    model_df = sub[[
        "OS_months", "event", "log_TMB",
        "Current Age", "Sex", "Race_clean"
    ]].dropna().copy()

    model_df = pd.get_dummies(
        model_df,
        columns=["Sex", "Race_clean"],
        drop_first=True
    )

    cph_sub = CoxPHFitter()
    cph_sub.fit(model_df, duration_col="OS_months", event_col="event")

    row = cph_sub.summary.loc["log_TMB", [
        "exp(coef)",
        "exp(coef) lower 95%",
        "exp(coef) upper 95%",
        "p"
    ]].to_dict()

    row["subgroup"] = subgroup
    row["n"] = model_df.shape[0]
    row["events"] = model_df["event"].sum()

    subgroup_results.append(row)

subgroup_df = pd.DataFrame(subgroup_results)


# 3. Order rows for plotting
order = ["Non-Stage IV", "Stage IV"]
subgroup_df["subgroup"] = pd.Categorical(subgroup_df["subgroup"], categories=order, ordered=True)
subgroup_df = subgroup_df.sort_values("subgroup").reset_index(drop=True)

# 4. Create labels
subgroup_df["label"] = subgroup_df.apply(
    lambda row: f'{row["subgroup"]}  (n={int(row["n"])}, events={int(row["events"])})',
    axis=1
)

# 5. Plot forest plot
plt.figure(figsize=(8, 4.2))
y_pos = np.arange(len(subgroup_df))

plt.errorbar(
    x=subgroup_df["exp(coef)"],
    y=y_pos,
    xerr=[
        subgroup_df["exp(coef)"] - subgroup_df["exp(coef) lower 95%"],
        subgroup_df["exp(coef) upper 95%"] - subgroup_df["exp(coef)"]
    ],
    fmt='o',
    color='black',
    ecolor='black',
    elinewidth=1.5,
    capsize=4,
    markersize=7
)

# Reference line
plt.axvline(1, color='black', linestyle='--', linewidth=1)

# Labels
plt.yticks(y_pos, subgroup_df["label"], fontsize=11)
plt.xlabel("Hazard Ratio for log(TMB)", fontsize=12)
plt.title("Effect of TMB on Survival by Disease Stage", fontsize=14)

# Use log scale for HR
plt.xscale("log")
plt.xlim(0.6, 1.25)

ticks = [0.6, 0.7, 0.8, 0.9, 1.0, 1.1, 1.2]
plt.xticks(ticks)

# Decimal lable
plt.gca().xaxis.set_major_formatter(mticker.FormatStrFormatter('%.1f'))

# Add HR text to the right
for i, row in subgroup_df.iterrows():
    hr = row["exp(coef)"]
    lower = row["exp(coef) lower 95%"]
    upper = row["exp(coef) upper 95%"]
    p = row["p"]

    plt.text(
        x=1.13,
        y=i,
        s=f"{hr:.2f} ({lower:.2f}–{upper:.2f}), p={p:.3g}",
        va="center",
        fontsize=10
    )

# Clean up style
ax = plt.gca()
for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)

plt.tight_layout()
plt.savefig("forest_subgroup.png", dpi=300)
plt.show()
```

The protective effect of TMB is observed only in non-metastatic patients, but disappears in Stage IV.
